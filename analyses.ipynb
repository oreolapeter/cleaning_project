```python

# Sql query 

query = """ SELECT DISTINCT name, d.entitled_bonus Stipends

FROM employees e

Left JOIN departments d

ON e.department_id = d._id

ORDER BY entitled_bonus DESC """

cursor.execute(query)

```

```python
# output the result of the query with .fetchall() method and assign it to a variable

employees_bonus = cursor.fetchall()

employees_bonus
```

```python
# create a dataframe of the result using pandas

employees_bonus_df = pd.DataFrame(employees_bonus,columns=['Name','Stipends'])

employees_bonus_df
```
